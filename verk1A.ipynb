```python
from kmeans import kmeans 
```

```python
%matplotlib inline
import kmeans
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage
```

```python
k = 3

# lesa in mynd
img = scipy.ndimage.imread('1.jpg', False, 'RGB')
print(img.shape)

# gera myndina(fylkið) tilbúna fyrir k-means
data = img.transpose(2,0,1).reshape(3,-1)
print(data.shape)

# nota k-means á fylkið
centroids, labels, J = kmeans.kmeans(data, k)

# sýna myndina
plt.imshow(img)
plt.show()

# sýna litina
plt.scatter([-0.02,0,0.02],[-0.04,0,-0.04], s=1000, c=np.transpose(centroids)/255)
plt.show

# sýna fjölda ítrana
print(len(J))
```
